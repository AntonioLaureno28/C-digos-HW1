### G(s) alpha = 8

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step

# Parâmetro alfa
alpha = 8

# Coeficientes do numerador de G(s): 5*(1 + alpha*s) = 5*alpha*s + 5
num_G_coeffs = [5*alpha, 5]

# Coeficientes dos termos do denominador de G(s)
# scipy.signal.TransferFunction espera os coeficientes do polinômio em ordem decrescente de potência.
# Termo (1 + 0.05s)^2 = (0.05s + 1)^2 = 0.0025s^2 + 0.1s + 1
# Coeficientes: [0.0025, 0.1, 1]
den_term1_poly = np.poly1d([0.05, 1])
den_term1_coeffs_sq = np.polymul(den_term1_poly, den_term1_poly)

# Termo (1 + 10s)^2 = (10s + 1)^2 = 100s^2 + 20s + 1
# Coeficientes: [100, 20, 1]
den_term2_poly = np.poly1d([10, 1]) 
den_term2_coeffs_sq = np.polymul(den_term2_poly, den_term2_poly)

# Termo (1 + 0.2s/10 + s^2/100) = (1/100)s^2 + (0.2/10)s + 1 = 0.01s^2 + 0.02s + 1
# Coeficientes: [0.01, 0.02, 1]
den_term3_coeffs = [1/100, 0.2/10, 1]

# Termo (1 + 0.4s/12 + s^2/144) = (1/144)s^2 + (0.4/12)s + 1
# Coeficientes: [1/144, 0.4/12, 1]
den_term4_coeffs = [1/144, 0.4/12, 1]

# Termo (1 + 1.6s/24 + s^2/576) = (1/576)s^2 + (1.6/24)s + 1
# Coeficientes: [1/576, 1.6/24, 1]
den_term5_coeffs = [1/576, 1.6/24, 1]

# Multiplicando todos os polinômios do denominador de G(s)
den_G_coeffs = den_term1_coeffs_sq 
den_G_coeffs = np.polymul(den_G_coeffs, den_term2_coeffs_sq)
den_G_coeffs = np.polymul(den_G_coeffs, den_term3_coeffs)
den_G_coeffs = np.polymul(den_G_coeffs, den_term4_coeffs)
den_G_coeffs = np.polymul(den_G_coeffs, den_term5_coeffs)

# Criando o objeto TransferFunction para G(s)
G_system = TransferFunction(num_G_coeffs, den_G_coeffs)

# Calculando a resposta ao degrau unitário para G(s)
# A função 'step' da scipy já considera uma entrada degrau unitário (U(s) = 1/s)
t, y = step(G_system)

# Plotagem do grafico
plt.plot(t, y)
plt.title(f"Resposta ao Degrau Unitário G(s) (alpha={alpha})")
plt.xlabel("Tempo [s]")
plt.ylabel("Saida y(t)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step

# Aproximação de primeira ordem: G(s) = 5 / (10s + 1)
num = [5]
den = [10, 1]

# Criar sistema
system = TransferFunction(num, den)

# Obter resposta ao degrau
t, y = step(system)

# Plotar
plt.figure(figsize=(8, 4))
plt.plot(t, y)
plt.title("Resposta ao Degrau da Aproximação de Primeira Ordem")
plt.xlabel("Tempo [s]")
plt.ylabel("Saída")
plt.grid(True)
plt.show()
```
