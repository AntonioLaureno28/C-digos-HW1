```python
def modo_1(t):
    return np.exp(-3*t)
def modo_2(t):
    return t*np.exp(-3*t)
```

```python
plt.plot(t,y_1)
plt.xlabel("Tempo(s)")
plt.ylabel("Modo 1")
plt.grid(True) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a fun√ß√£o para a evolu√ß√£o livre y(t)
def y_fe(t):
    return 3 * np.exp(-3 * t) + 10 * t * np.exp(-3 * t)

# Intervalo de tempo
t = np.linspace(0, 10,1000) 

# Calculando os valores de y(t) 
y = y_fe(t)

# Plotando o gr√°fico
plt.plot(t, y,)
plt.title('Evolu√ß√£o Livre do Sistema')
plt.xlabel('Tempo (s)')
plt.ylabel('y(t)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Definindo os coeficientes do numerador e do denominador de G(s)
num_G = [2, 15, 31]
den_G = [1, 6, 9]

#Criando o objeto TransferFunction usando scipy
G_system = signal.TransferFunction(num_G, den_G)

# Calculando a resposta ao degrau
t = np.linspace(0, 5, 1000)
t_step, y_step = signal.step(G_system, T=t)

#Plotando a resposta ao degrau
plt.plot(t_step, y_step)
plt.title("Resposta ao degrau")
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude $y(t)$')
plt.grid(True)
plt.show() 
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Definindo os coeficientes do numerador e do denominador de G(s)
num_G = [2, 15, 31]
den_G = [1, 6, 9]

#Criando o objeto TransferFunction usando scipy
G_system = signal.TransferFunction(num_G, den_G)

# Calculando a resposta ao impulso
t = np.linspace(0, 5, 500)
t_impulse, y_impulse = signal.impulse(G_system, T=t)

#Plotando a resposta ao impulso
plt.plot(t_impulse, y_impulse)
plt.title("Resposta ao impulso")
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude $y(t)$')
plt.grid(True)
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step

# Coeficientes do numerador expandido com alpha = 8
alpha = 8
num = np.polymul([1, alpha], [0.001736, 0.08333, 1])
num = np.polymul(num, [0.006944, 0.16667, 1])
num = np.polymul(num, [5])  # Constante multiplicadora

# Coeficientes do denominador
den1 = np.poly1d([0.05, 1])**2
den2 = np.poly1d([10, 1])**2
den3 = [0.01, 0.2, 1]
den = np.polymul(np.polymul(den1, den2), den3)
den = np.polymul(den, [1, 0])  # Multiplica por s para entrada degrau

# Criando o sistema
system = TransferFunction(num, den)

# Resposta ao degrau
t, y = step(system)

# Plotagem do gr√°fico
plt.plot(t, y)
plt.title("Resposta ao Degrau Unitario(alpha= 8)")
plt.xlabel("Tempo [s]")
plt.ylabel("Saida")
plt.grid()
plt.show()
```
