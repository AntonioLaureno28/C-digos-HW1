```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def modo_1(t):
    return np.exp(-3*t)
def modo_2(t):
    return t*np.exp(-3*t)
```

```python
t = np.linspace(0,5,20)
y_1 = modo_1(t)
y_2 = modo_2(t)
```

```python
plt.plot(t,y_1)
plt.xlabel("Tempo(s)")
plt.ylabel("Modo 1")
```

```python
plt.plot(t,y_2)
plt.xlabel("Tempo(s)")
plt.ylabel("Modo 2")
```
