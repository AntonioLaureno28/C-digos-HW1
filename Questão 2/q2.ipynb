```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def modo_1(t):
    return np.exp(-3*t)
def modo_2(t):
    return t*np.exp(-3*t)
```

```python
t = np.linspace(0,5,2000)
y_1 = modo_1(t)
y_2 = modo_2(t)
```

```python
plt.plot(t,y_1)
plt.xlabel("Tempo(s)")
plt.ylabel("Modo 1")
plt.grid(True) 
plt.show()
```

```python
plt.plot(t,y_2)
plt.xlabel("Tempo(s)")
plt.ylabel("Modo 2")
plt.grid(True) 
plt.show()
```

```python
from scipy import signal

#Defini√ß√£o do sistema G(s) para a resposta ao impulso h(t))
num_G = [2, 15, 31]
den_G = [1, 6, 9]
G_system = signal.TransferFunction(num_G, den_G)

#Coeficientes para Y_fe(s)
num_Yfe = [3, 19]
den_Yfe = [1, 6, 9]
Yfe_system = signal.TransferFunction(num_Yfe, den_Yfe)

#Vetor de tempo 
t = np.linspace(0, 5, 1000) 

#Calcula a evolu√ß√£o livre y_fe(t) usando signal.
t_out_fe, y_fe = signal.impulse(Yfe_system, T=t)

# Calcula a parte cont√≠nua da resposta ao impulso h(t) do sistema G(s)
t_out_h, Hc_values = signal.impulse(G_system, T=t)

# Soma as duas respostas cont√≠nuas
y = y_fe + Hc_values

# Plotagem do gr√°fico
plt.plot(t, y)
plt.title('Resposta completa') 
plt.ylabel('$y(t)$')
plt.grid(True)
plt.show()
```
